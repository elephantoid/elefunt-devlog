- 논리 연산자 `>`, `<`, `==`

```python
df_Homelessness ['Count'] > 500
```

```python
df_Homelessness[df_Homelessness['Count'] > 500]
```

```python
df_Homelessness[df_Homelessness['State'] == 'WY']
```

```python
#hide
df_Homelessness['Year']=pd.to_datetime(df_Homelessness['Year'])
```

```python
df_Homelessness[df_Homelessness['Year'] > "2015-01-01"]
```

```python
is_AK= df_Homelessness['State'] =='AK'
is_cnt0= df_Homelessness['Count'] == 0
# State = Ak and Count=0 인 행
df_Homelessness[is_AK & is_cnt0][:10]
```

```python
# 위 와 동일함 각 조건에 ()추가
df_Homelessness[ (df_Homelessness['State']== 'AK') & (df_Homelessness['Count']==0)][:10]
```
